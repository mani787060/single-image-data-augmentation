```python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('cat pic.jpg', target_size=(200,200))

img
```

```python
type(img)
```

```python
# introducing 'data_augmentation' for the given figure

datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)   # shows that we take 1 image of this size.
```

```python
i=0

# we have to make a directory named 'day50_aug' to see the changes-
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='day50_aug'):

    i = i + 1

    if  i == 10:
        break
        
```
